# Multistart Algorithm

Crea istanza

```python
from utils import *

NUM_MACHINES = 8
NUM_JOBS = 100
MAX_TIME_JOBS = 100
SORT_FROM_LONGEST = True

instance_problem = create_instance_problem(NUM_JOBS,MAX_TIME_JOBS,NUM_MACHINES,SORT_FROM_LONGEST)

print(instance_problem["list_jobs"])
```

Da fare

```python
initial_solutions = []
best_solutions = []

for i in range(10):
    initial_solutions.append(greedy_algorithm(instance_problem)[0])
    sol = local_search_algorithm(initial_solutions[i],neighborWay=swap_move)[0]
    best_solutions.append({"schedule":sol,"result":get_solution(sol)})

best_solution_finded = min(best_solutions, key=lambda s:s["result"])
print("best time solution finded: ", best_solution_finded["schedule"])
```
