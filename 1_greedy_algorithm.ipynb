# Parallelal machine scheduling

#### Create instance of problem

```python
from utils import *

NUM_MACHINES = 8
NUM_JOBS = 100
MAX_TIME_JOBS = 1000
SORT_FROM_LONGEST = True

instance_problem = create_instance_problem(NUM_JOBS,MAX_TIME_JOBS,NUM_MACHINES,SORT_FROM_LONGEST)

print(instance_problem["list_jobs"])
```

#### Deterministic algorithm
Choose the longest/shortest job at each iteration

```python
schedule,time_execution_function = greedy_algorithm(instance_problem, deterministic_way)

print(f"SCHEDULE: {schedule}")

solution = get_solution(schedule)

print(f"\nSOLUTION: {solution}")
```

#### Non-deterministic algorithm
Randomly chooses among the top 10 longest/shortest jobs at each iteration

```python
schedules: list[list[MachineType]] =[]
solutions = []

for i in range(10):
    schedules.append(greedy_algorithm(instance_problem,non_deterministic_way)[0])
    print(f"SCHEDULE {i}: {schedules[i]}")
    solutions.append(get_solution(schedules[i]))
    print(f"Solution {solutions[i]}\n")

print(f"\nBEST SOLUTION: {min(solutions)}")
```

#### Totally random algorithm
Randomly chooses among all jobs at each iteration

```python
schedules_random: list[list[MachineType]] =[]
solutions_by_random_algorithm = []

for i in range(10):
    schedules_random.append(greedy_algorithm(instance_problem,random_way)[0])
    print(f"SCHEDULE {i}: {schedules_random[i]}")
    solutions_by_random_algorithm.append(get_solution(schedules_random[i]))
    print(f"Solution: {solutions_by_random_algorithm[i]}\n")

print(f"\nBEST TOTAL TIME: {min(solutions_by_random_algorithm)}")
```
